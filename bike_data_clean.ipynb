```python
import pandas as pd
import numpy as np
```

```python
bike_data = pd.read_csv("Resources/citibikedata_sixmo.csv")

bike_data.head()
```

```python
bike_data.isnull().sum()
```

```python
bike_data = bike_data.dropna()

bike_data.isnull().sum()
```

```python
bike_data.dtypes
```

```python
dupes = bike_data.duplicated(subset=["ride_id"])

dupe_list = list(np.where(dupes)[0])

len(dupe_list)
```

```python
bike_data["started_at"] = pd.to_datetime(bike_data["started_at"], format='%Y-%m-%d %H:%M:%S')
bike_data["ended_at"] = pd.to_datetime(bike_data["ended_at"], format='%Y-%m-%d %H:%M:%S')

bike_data.head()
```

```python
bike_data.dtypes
```

```python
bike_data["ride_duration"] = round((bike_data["ended_at"] - bike_data["started_at"]).dt.seconds / 60, 2)

bike_data.head()
```

```python
len(bike_data['end_station_id'].unique())
```

```python
len(bike_data["start_station_id"].unique())
```

```python
endcoords = {}

for index, row in bike_data.iterrows():
    
    end_id = row['end_station_id']
    
    if end_id not in endcoords:
        
        endcoords[end_id] = row["end_lat"], row["end_lng"]
        
print(endcoords)
```

```python
coords = {}

for index, row in bike_data.iterrows():
    
    start_id = row['start_station_id']
    
    if start_id not in endcoords.keys() and start_id not in coords:
        
        coords[start_id] = row["start_lat"], row["start_lng"]
        
print(coords)
```

```python
endcoords.update(coords)

endcoords
```

```python
for index, row in bike_data.iterrows():
    
    check = row['start_station_id']
    
    if check in endcoords.keys():
        
        bike_data.at[index, 'start_lat'] = endcoords[check][0]
        bike_data.at[index, 'start_lng'] = endcoords[check][1]

bike_data.head()
```

```python
bike_data["start_lat"].value_counts()
```

```python
bike_data["start_lng"].value_counts()
```

```python
def haversine(lat1, lon1, lat2, lon2, to_radians=True, earth_radius=6371):

    if to_radians:
        lat1, lon1, lat2, lon2 = np.radians([lat1, lon1, lat2, lon2])

    a = np.sin((lat2-lat1)/2.0)**2 + \
        np.cos(lat1) * np.cos(lat2) * np.sin((lon2-lon1)/2.0)**2

    return earth_radius * 2 * np.arcsin(np.sqrt(a))


#df['dist'] = \
#    haversine(df.LAT.shift(), df.LONG.shift(),
#                 df.loc[1:, 'LAT'], df.loc[1:, 'LONG'])

bike_data["distance_traveled"] = haversine(bike_data["start_lat"], bike_data["start_lng"], bike_data["end_lat"], bike_data["end_lng"]) * 0.621371

bike_data["distance_traveled"] = round(bike_data["distance_traveled"], 2)

bike_data.head()
```

```python
bike_data.to_csv("Resources/citibikedata_final.csv", encoding='utf8', index=False)
```
